```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font' , family=font_name);

```

```python
one = pd.read_csv('../data/0104.csv', parse_dates=['일자'], encoding='euc-kr')
two = pd.read_csv('../data/0508.csv', parse_dates=['일자'], encoding='euc-kr')
SEP = pd.read_csv('../data/CALL_HOSP_09MONTH.csv', parse_dates=['일자'])
OCT = pd.read_csv('../data/CALL_HOSP_10MONTH.csv', parse_dates=['일자'])
NOV = pd.read_csv('../data/CALL_HOSP_11MONTH.csv', parse_dates=['일자'])
DEC = pd.read_csv('../data/CALL_HOSP_12MONTH.csv', parse_dates=['일자'])
```

# 통화건수

```python
x = pd.concat([one,two,SEP.loc[SEP["발신지_시도"] == "서울특별시"],OCT.loc[OCT["발신지_시도"] == "서울특별시"],NOV.loc[NOV["발신지_시도"] == "서울특별시"],DEC.loc[DEC["발신지_시도"] == "서울특별시"]]) 
print(x)
```
